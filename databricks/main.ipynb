```python
import json
import pandas as pd
print("Bibliotecas ok !")
```

```python
with open('../data/amazon-data.json', 'r') as f:
    data = json.load(f)

df = pd.json_normalize(data['data']['deals'],sep=".")
```

```python
df_dim = df[[
    'product_asin',
    'deal_type',
    'deal_title',
    'deal_state',
    'deal_url',
    'deal_photo'
]]

df_dim
```

```python
df_fact = df[[
    'deal_id',
    'product_asin',
    'deal_price.amount',
    'list_price.amount',
    'savings_amount.amount',
    'deal_price.currency',
    'savings_percentage',
    'deal_starts_at',
    'deal_ends_at',
]]
df_fact
```

```python
df_fact = df_fact.rename(columns={
    'deal_price.amount': 'deal_price',
    'list_price.amount': 'list_price',
    'savings_amount.amount': 'savings_amount',
    'deal_price.currency': 'deal_price_currency',
})

num_cols = ['deal_price', 'list_price', 'savings_amount']
tstp_cols =  ['deal_starts_at', 'deal_ends_at']

for col in num_cols:
    df_fact[col] = pd.to_numeric(df_fact[col])

for col in tstp_cols:
    df_fact[col] = pd.to_datetime(df_fact[col])
```

```python
# Check por duplicidades
print("Número de duplicados em df_dim:", df_dim.duplicated().sum())
print("Número de duplicados em df_fact:", df_fact.duplicated().sum())

# Contagem de linhas
print("Número de linhas em df_dim:", len(df_dim))
print("Número de linhas em df_fact:", len(df_fact))
```

```python
df_join = df_fact.merge(df_dim, on='product_asin', how='left')
df_join
```
